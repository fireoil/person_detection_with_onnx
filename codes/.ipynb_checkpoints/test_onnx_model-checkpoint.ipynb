# 人脸检测

```python
import cv2
import onnxruntime as ort
import argparse
import numpy as np
from dependencies.box_utils import predict

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# scale current rectangle to box
def scale(box):
    width = box[2] - box[0]
    height = box[3] - box[1]
    maximum = max(width, height)
    dx = int((maximum - width)/2)
    dy = int((maximum - height)/2)
    
    bboxes = [box[0] - dx, box[1] - dy, box[2] + dx, box[3] + dy]
    return bboxes

# crop image
def cropImage(image, box):
    num = image[box[1]:box[3], box[0]:box[2]]
    return num

# face detection method
def faceDetector(orig_image, threshold = 0.7):
#     orig_image = cv2.imread(img_path)
    image = cv2.resize(orig_image, (640, 480))
    image_mean = np.array([127, 127, 127])
    image = (image - image_mean) / 128
    image = np.transpose(image, [2, 0, 1])
    image = np.expand_dims(image, axis=0)
    image = image.astype(np.float32)

    input_name = face_detector.get_inputs()[0].name
    confidences, boxes = face_detector.run(None, {input_name: image})
    boxes, labels, probs = predict(orig_image.shape[1], orig_image.shape[0], confidences, boxes, threshold)
    return boxes, labels, probs
```

```python
import os.path as osp
import cv2

def crop_box(img, box):
    return img[box[1] : box[3],  box[0] :  box[2]]

def crop_boxes(img, boxes):
    imgs = [None]*len(boxes)
    for idx, box in enumerate(boxes):
        imgs[idx] = crop_box(img, box)
    return imgs

def save_crop_boxes(img_path, img, boxes):
    len_boxes = len(boxes)
    filename = img_path.split("/")[-1]
    filename_, expansion = filename.split(".")[0], filename.split(".")[-1]
    faces = crop_boxes(img, boxes)
    filenames = [filename_ + "_" + str(idx) + "." + expansion for idx in range(len_boxes)]
    filepaths = [osp.join("./faces", filename) for filename in filenames]
    for filepath, face in zip(filepaths, faces):
        cv2.imwrite(filepath, np.array(face[:, :, ::-1]))
```

```python
def detect_face(img_path):
    orig_image = cv2.imread(img_path)
    orig_image = cv2.cvtColor(orig_image, cv2.COLOR_BGR2RGB)
    boxes, labels, probs = faceDetector(orig_image)
    
#     save_crop_boxes(img_path, orig_image, boxes)
    
    plt.figure(figsize=(10, 10))
    for i in range(boxes.shape[0]):
        box = scale(boxes[i, :])
        cv2.rectangle(orig_image, (box[0], box[1]), (box[2], box[3]), (255, 128, 0), 4)
        plt.imshow(orig_image)
```

```python
def detect_single_face(img_path):
    orig_image = cv2.imread(img_path)
    orig_image = cv2.cvtColor(orig_image, cv2.COLOR_BGR2RGB)
    boxes, labels, probs = faceDetector(orig_image)
    
    face = crop_box(orig_image, boxes[0])
    return face

def show_single_face(face):
    plt.imshow(face)
    plt.axis("off")
    plt.show()
```

# 人脸识别

```python
import numpy as np
import onnx
import onnxruntime as ort
```

```python
def get_onnx_input_tensor(img_path):
    orig_image = cv2.imread(img_path, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(orig_image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (112, 112))
    image_mean = np.array([127, 127, 127])
    image = (image - image_mean) / 128
    image = np.transpose(image, [2, 0, 1])
    image = np.expand_dims(image, axis=0)
    image = image.astype(np.float32)
    return image
```

```python
def get_onnx_input_tensor_from_np(img_np):
    img_np = np.array(img_np)
    image = cv2.resize(img_np, (112, 112))
    image_mean = np.array([127, 127, 127])
    image = (image - image_mean) / 128
    image = np.transpose(image, [2, 0, 1])
    image = np.expand_dims(image, axis=0)
    image = image.astype(np.float32)
    return image
```

```python
def test_onnx_img(onnx_path, img_np):
    sess = ort.InferenceSession(onnx_path)
    input_name = sess.get_inputs()[0].name
    output_name = sess.get_outputs()[0].name
    print(input_name, output_name)
    ort_data = {input_name: get_onnx_input_tensor_from_np(img_np)}
    output = sess.run([output_name], ort_data)
    return output
```

```python
def cosine_distance(a, b):
    if a.shape != b.shape:
        raise RuntimeError("array {} shape not match {}".format(a.shape, b.shape))
    if a.ndim==1:
        a_norm = np.linalg.norm(a)
        b_norm = np.linalg.norm(b)
    elif a.ndim==2:
        a_norm = np.linalg.norm(a, axis=1, keepdims=True)
        b_norm = np.linalg.norm(b, axis=1, keepdims=True)
    else:
        raise RuntimeError("array dimensions {} not right".format(a.ndim))
    similiarity = np.dot(a, b.T)/(a_norm * b_norm) 
    return similiarity
```
